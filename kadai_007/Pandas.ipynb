```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")

# 単価に消費税をかけた値を返す
def tax(x):
    return x * 1.10

# 単価にtaxをかける
df['単価'].apply(tax)
```

```python
tax_series = df['単価'].apply(tax)
tax_series.name = "単価（税込み）"
pd.concat([df, tax_series], axis=1)
```

```python
category_df = pd.read_csv('category.csv')
category_df
```

```python
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df
```

```python
# 必要なライブラリをインポート
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

# カテゴリーのユニーク出現頻度のカウント
num = df['カテゴリー'].value_counts()

# カテゴリー別、ユニーク出現頻度の棒グラフ表示
num.plot(kind='bar')

# タイトル＆各ラベルの設定
plt.xlabel('カテゴリー')
plt.xticks(rotation='horizontal')
plt.ylabel('出現頻度', rotation='horizontal')
plt.title('カテゴリーの出現頻度')

# グラフを表示
plt.show()

# データフレームを基準列ごとに要素をグループにまとめ、サブ列を集計するメソッド
print(df.groupby('商品番号')['カテゴリー'].value_counts())

# データフレームの基本統計量を算出するメソッド
print(df.groupby('商品番号')['カテゴリー'].describe())
```
